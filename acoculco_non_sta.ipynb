### Modelo conceptual

Acoculco es un sistema geotérmico que se encuentra en la Faja Volcánica Trans-mexicana, al oriente de esta, en el estado de Puebla, México.

Acoculco (pozo EAC-1) muestran un régimen de transferencia de calor conductivo conductiva bajo un gradiente geotérmico elevado (López-Hernández et al., 2009), es así, que se tiene como propósito determinar el tiempo en que ocurrió el cambio de régimen de transferencia de calor de uno convectivo a uno conductivo.
Se desarrolla un modelo 1D que simula la distribución tiempo-profundidad de la temperatura. Las curvas de salida se comparan con los datos experimentales proporcionados por las inclusiones de fluidos y por el registro térmico de la perforación. 

Las características del modelo conceptual son las siguientes:

1. Régimen de transferencia de calor por conducción. (De convección a conducción)
2. Se supone una estructura estratificada subterránea, en donde cada capa es isótropa y tiene propiedades constantes.
2. Uso del método de diferencias finitas.
3. Se utiliza, para ello, la resolución de una ecuación por balance de calor.
4. En el modelo, no se consideran fuentes ni sumideros.
6. Las temperaturas inicial y final son fijas en los extremos (condiciones de frontera).

### Modelo matemático (Caso no estacionario)

La ecuación de balance de calor en 1D, se describe como sigue:

$ \frac{\partial T}{\partial t} = D_{th} \frac{\partial T^2}{\partial z^2}$

$D_{th} = \frac{\kappa}{\rho c}$

Donde: 
$D_{th}$ = Difusividad térmica [$10^{-7} \frac{m^2}{s}$]

#### Coeficientes y parámetros físicos

$T$ = Temperatura [°C]

$z$ = Profundidad [m]

$t$ = Tiempo [años]

$\kappa$ = Conductividad térmica [$\frac{W}{m  ^°C}$]

$\rho$ = Densidad de masa [$\frac{kg}{m^3}$]

$c$ = capcaidad térmica específica [$\frac{J}{kg K}$]

Se considera la ec. de Poisson con $k$ variable y condiciones de frontera tipo Dirichlet:

$g=k{\frac{du}{dx}}$

${\frac{d}{dx}}(k{\frac{du}{dx}}) = \frac{dg}{dx}$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import interpolate
from fdm_1D import calcDth, harmonicMean, arithmeticMean, Laplaciano1D_NS, interpTemp
# Parámetros para el estilo de las gráficas
# Si tienes instalado LaTeX, lo siguiente hace los textos 
# más estilizados. Si no tienes LaTeX, debes comentar la 
# línea: 'text.usetex': True
plt.style.use('seaborn-paper')
params = {'figure.figsize' : (14,7),
          'text.usetex'    : True,
          'xtick.labelsize': 20,
          'ytick.labelsize': 20,
          'axes.labelsize' : 24,
          'axes.titlesize' : 24,
          'legend.fontsize': 24,
          'lines.linewidth': 3,
          'lines.markersize': 10,
          'grid.color'     : 'darkgray',
          'grid.linewidth' : 0.5,
          'grid.linestyle' : '--',
          'font.family': 'DejaVu Serif',
         }
plt.rcParams.update(params)
```

```python
#
# Datos físicos del problema
#
Ttop = 15   # Temperatura en la superficie
Tbot = 750  # Temperatura a 4000 m
L = 4000.0  # Profundidad máxima
```

```python
z= np.asarray([0, 50, 250, 400, 600, 800, 1000, 1500, 1900])
z2 = np.array([50, 250, 400, 600, 800, 1000, 1500, 1900, 4000])
k= np.asarray([1.84,1.41,1.36,1.44,1.46,1.68,1.84,1.66,2.10])
c= np.asarray([840, 840, 840, 840, 840, 840, 890, 890, 950])
p= np.asarray([2120, 2340, 2300, 2300, 2290, 2700, 2700, 2360, 2760])

print('z = {}'.format(z))
print('z2 = {}'.format(z2))
print('k = {}'.format(k))
print('c = {}'.format(c))
print('p = {}'.format(p))
```

```python
Tabla = pd.DataFrame(np.array([z, z2, k, c, p]).transpose(), 
                     columns=['$z$', '$z_2$','$k$','$c$','$p$'])
Tabla
```

```python
#Cálculo del coeficiente Dth
Tabla["Dth"] = k/(p*c) 
DthData = k/(p*c)
Tabla
```

```python
# Datos para la solución numérica por diferencias finitas
#
N = 99                      # Número de incógnitas
dx = L / (N+1)              # Distancia entre nodos
z = np.linspace(0, L, N+2)  # Coordenada z de la profundidad
dt = 3600*24*365 # Paso de tiempo (seconds per year)
NT = 250000      # Número total de pasos de tiempo
```

```python
print('\nDatos para la solución numérica usando FD')
print(' dx = {:10.1f} [m] \n L  = {:10.1f} [m] \n N  = {:8d}'.format(dx, L, N))
print(' dt = {:10.1f} [s] \n NT = {:8d}'.format(dt, NT))
```

```python
# Aceleración para el método implícito
#
SP = 1000
dt *= SP
NT /= SP
print('\nUsamos un factor de aceleración debido al método implícito: ')
print(' SP = {:13d} (factor)\n dt = {:15.1f} [s] \n NT = {:13d} (pasos totales)'.format(SP, dt, int(NT)))
print('\nz shape = {} \n{}'.format(z.shape,z))
#input('Presiona <enter>')
#
```

```python
A = calcDth(DthData, z, N)
print('\nCoeficiente en los puntos de la simulación: \n D_th shape = {} \n{}'.format(A.shape, A))
```

```python
#
# SOLUCIÓN USANDO DIFERENCIAS FINITAS
#
# Condición inicial: Profundidades
z_dat = [0, 100, 200, 400, 710, 803, 1100, 1200, 1400, 1500, 1600, 1700, 1800, 2000, 2500, 3000, 3500, 4000]
# Condición inicial: Temperaturas
T_dat = [15, 113, 145, 178, 155, 201, 215, 282, 223, 226, 252, 284, 310, 350, 450, 550, 650, 750]
# Interpolación a los lugares de la simulación
tck_1 = interpolate.splrep(z_dat, T_dat, s = 0)
T_ini = interpolate.splev(z, tck_1, der = 0)
#
# Temperaturas actuales
z_dat_now = [0, 96, 192, 389, 596, 798, 1000, 1197, 1408, 1495, 1586, 1649, 1702, 1750, 1803, 1851, 1904, 1942, 1967, 2000, 2500, 3000, 3500, 4000]
T_dat_now = [15, 69, 74, 95, 121, 159, 168, 192, 213, 213, 236, 251, 267, 272, 284, 288, 298, 293, 310, 325, 440, 550, 650, 750]
# Interpolación a los lugares de la simulación
tck_2 = interpolate.splrep(z_dat_now, T_dat_now, s = 0)
T_final = interpolate.splev(z, tck_2, der = 0)
#
# SOLUCIÓN USANDO DIFERENCIAS FINITAS
#
# Definicion del sistema lineal de N x N
#
dt_dx2 = dt / dx**2
gamma = A * dt_dx2
A2 = Laplaciano1D_NS(N, gamma) # Matriz del sistema

print('\nGamma shape = {}, A shape = {}'.format(gamma.shape, A.shape))
#
# Ciclo de solución temporal
#
T_new = T_ini.copy()
T_now = np.zeros(N+2)
for i in range(0, int(NT)+1):
    print(i)
    T_new[ 1] += gamma[0] * T_ini[0]
    T_new[-2] += gamma[-1] * T_ini[-1]    
    T_new[1:-1] = np.linalg.solve(A2,T_new[1:-1])
    if (i == 7):
        T_now = T_new.copy() # Temperatura actual
#
```

```python
# GRAFICACIÓN
#
ax1 = plt.subplot(121)
plt.plot([T_ini[0], T_ini[-1]], [z[0], z[-1]], '--', lw=0.75, c='k' )
plt.scatter(T_dat, z_dat, s=20, alpha=0.5)
plt.plot(T_ini, z, '-', lw = 1.5, label='Cond. inicial')
plt.scatter(T_dat_now, z_dat_now, s=20, alpha=0.5, label='Datos de pozo')
plt.plot(T_now, z, label='Simulación Temp. actual')
plt.plot(T_new,z, label='Simulación Temp. final')
plt.ylim(4100,-100)
plt.xlim(-20,800)
plt.title('Solución Numérica')
plt.xlabel('$T$ [$^o$C]')
plt.ylabel('Profundidad [m]')
plt.grid()
plt.legend()

ax2 = plt.subplot(122, sharey=ax1)
plt.plot(A, z, '.--', lw=0.75, c='k')
plt.xlabel('$D_{th}$ [m$^2$/s]')
plt.xlim(0.65e-6,1.05e-6)
plt.minorticks_on()
plt.ticklabel_format(axis='x', style='sci', scilimits=(0,0))
plt.title('Coeficiente $D_{th}$')
plt.grid()

plt.subplots_adjust(hspace=1.0)

plt.show()
```
