### Modelo conceptual

Acoculco es un sistema geotérmico que se encuentra en la Faja Volcánica Trans-mexicana, al oriente de esta, en el estado de Puebla, México.

Acoculco (pozo EAC-1) muestran un régimen de transferencia de calor conductivo conductiva bajo un gradiente geotérmico elevado (López-Hernández et al., 2009), es así, que se tiene como propósito determinar el tiempo en que ocurrió el cambio de régimen de transferencia de calor de uno convectivo a uno conductivo.
Se desarrolla un modelo 1D que simula la distribución tiempo-profundidad de la temperatura. Las curvas de salida se comparan con los datos experimentales proporcionados por las inclusiones de fluidos y por el registro térmico de la perforación. 

Las características del modelo conceptual son las siguientes:

1. Régimen de transferencia de calor por conducción. (De convección a conducción)
2. Se supone una estructura estratificada subterránea, en donde cada capa es isótropa y tiene propiedades constantes.
2. Uso del método de diferencias finitas.
3. Se utiliza, para ello, la resolución de una ecuación por balance de calor.
4. En el modelo, no se consideran fuentes ni sumideros.
6. Las temperaturas inicial y final son fijas en los extremos (condiciones de frontera).

**Profundidad y características del pozo:**

### Modelo matemático (Caso estacionario)

La ecuación de balance de calor en 1D, se describe como sigue:

$\frac{\partial{T}}{\partial{t}} = D_{th} \frac{\partial{T^2}}{\partial{z^2}}$

En donde $\frac{\partial{T}}{\partial{t}}$ = 0 y representa la difusividad térmica

entonces,

$ D_{th} \frac{\partial{T^2}}{\partial{z^2}} = 0$

$ D_{th}$ se define como:

$ D_{th}= \frac{\kappa}{\rho c}$ [$10^{-7} \frac{m^2}{s}$]

#### Coeficientes y parámetros físicos

$T = $ temperatura [°C]

$z = $ profundidad [m]

$\kappa = $ conductividad térmica [$\frac{W}{m  ^°C}$]

$c = $ capacidad calorífica específica [$\frac{J}{kg K}$]

$\rho = $ densidad de masa [$\frac{kg}{m^3}$]

### Modelo numérico

Se considera la ec. de Poisson con $k$ variable y condiciones de frontera tipo Dirichlet:

$\frac{d}{dx}(k{\frac{du}{dx}})=f$ con $k=k(x)$

$g=k{\frac{du}{dx}}$

${\frac{d}{dx}}(k{\frac{du}{dx}}) = \frac{dg}{dx}$

$\frac{du}{dx}={\frac{k_{i+\frac{1}{2}}-k_{i-\frac{1}{2}}}{h}}$ (en i)

$k\frac{du}{dx}=k_{i+\frac{1}{2}} {{\frac{u_{i+1}-u_i}{h}}}$ (en $i+\frac{1}{2}$)

$k\frac{du}{dx}=k_{i+\frac{1}{2}} {{\frac{-u_{i-1}+u_i}{h}}}$ (en $i-\frac{1}{2}$)

### Modelo computacional (Caso estacionario)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from fdm_1D import calcDth, harmonicMean, arithmeticMean, Laplaciano1D
# Parámetros para el estilo de las gráficas
# Si tienes instalado LaTeX, lo siguiente hace los textos 
# más estilizados. Si no tienes LaTeX, debes comentar la 
# línea: 'text.usetex': True
plt.style.use('seaborn-paper')
params = {'figure.figsize' : (14,7),
          'text.usetex'    : True,
          'xtick.labelsize': 20,
          'ytick.labelsize': 20,
          'axes.labelsize' : 24,
          'axes.titlesize' : 24,
          'legend.fontsize': 24,
          'lines.linewidth': 3,
          'lines.markersize': 10,
          'grid.color'     : 'darkgray',
          'grid.linewidth' : 0.5,
          'grid.linestyle' : '--',
          'font.family': 'DejaVu Serif',
         }
plt.rcParams.update(params)
```

```python
# Datos físicos del problema
#
Ttop = 15   # Temperatura en la superficie
Tbot = 750  # Temperatura a 4000 m
L = 4000.0  # Profundidad máxima
```

```python
z= np.asarray([0, 50, 250, 400, 600, 800, 1000, 1500, 1900])
z2 = np.array([50, 250, 400, 600, 800, 1000, 1500, 1900, 4000])
k= np.asarray([1.84,1.41,1.36,1.44,1.46,1.68,1.84,1.66,2.10])
c= np.asarray([840, 840, 840, 840, 840, 840, 890, 890, 950])
p= np.asarray([2120, 2340, 2300, 2300, 2290, 2700, 2700, 2360, 2760])

print('z = {}'.format(z))
print('z2 = {}'.format(z2))
print('k = {}'.format(k))
print('c = {}'.format(c))
print('p = {}'.format(p))
```

```python
Tabla = pd.DataFrame(np.array([z, z2, k, c, p]).transpose(), 
                     columns=['$z$', '$z_2$','$k$','$c$','$p$'])
Tabla
```

```python
#Cálculo del coeficiente Dth
Tabla["Dth"] = k/(p*c) 
DthData = k/(p*c)
Tabla
```

```python
# Datos para la solución numérica por diferencias finitas
#
N1 = 99                      # Número de incógnitas
N2 = 95                     # Número de incógnitas para valorar el rango de nodos para un resultado correcto
N3 = 102
N4 = 110

dx = L / (N1+1)              # Distancia entre nodos
dx2 = L / (N2+1)
dx3 = L / (N3+1)
dx4= L / (N4+1)

z1 = np.linspace(0, L, N1+2)  # Coordenada z de la profundidad
z2 = np.linspace(0, L, N2+2)
z3 = np.linspace(0, L, N3+2)
z4 = np.linspace(0, L, N4+2)
```

```python
A = calcDth(DthData, z1, N1)
print('\nCoeficiente en los puntos de la simulación: \n D_th shape = {} \n{}'.format(A.shape, A))
B = calcDth(DthData, z2, N2)
C = calcDth(DthData, z3, N3)
D = calcDth(DthData, z4, N4)
```

```python
# SOLUCIÓN USANDO DIFERENCIAS FINITAS
#
# Definición del sistema lineal de N x N
#
gamma = A / dx**2
f = np.zeros(N1)            # RHS
A1 = Laplaciano1D(N1, gamma) # Matriz del sistema

print('\nGamma shape = {} \nf shape = {}, A shape = {}'.format(gamma.shape, f.shape, A.shape))
#
# Con N2 = 95 nodos
gamma2 = B / dx2**2
f2 = np.zeros(N2)            # RHS
B2 = Laplaciano1D(N2, gamma2) # Matriz del sistema

# Con N2 = 102 nodos
gamma3 = C / dx3**2
f3 = np.zeros(N3)            # RHS
C2 = Laplaciano1D(N3, gamma3) # Matriz del sistema

# Con N2 = 110 nodos
gamma4 = D / dx4**2
f4 = np.zeros(N4)            # RHS
D2 = Laplaciano1D(N4, gamma4) # Matriz del sistema
```

```python
# Arreglo para la solución
#
T = np.zeros(N1+2)
T2 = np.zeros(N2+2)
T3 = np.zeros(N3+2)
T4 = np.zeros(N4+2)
#
```

```python
# Aplicación de las condiciones de frontera tipo Dirichlet.
#
T[0] = Ttop
T[N1+1] = Tbot 
f[0] -= Ttop * gamma[0]
f[N1-1] -= Tbot * gamma[N1+1]
#
# Solución del sistema lineal
#
T[1:N1+1] = np.linalg.solve(A1,f)
#
print('\nSolución T shape = {} \n{}'.format(T.shape, T))
#

# Para N2 = 95 nodos
# Aplicación de las condiciones de frontera tipo Dirichlet.
#
T2[0] = Ttop
T2[N2+1] = Tbot 
f2[0] -= Ttop * gamma2[0]
f2[N2-1] -= Tbot * gamma2[N2+1]
#
# Solución del sistema lineal
#
T2[1:N2+1] = np.linalg.solve(B2,f2)
#

# Para N2 = 102 nodos
# Aplicación de las condiciones de frontera tipo Dirichlet.
#
T3[0] = Ttop
T3[N3+1] = Tbot 
f3[0] -= Ttop * gamma3[0]
f3[N3-1] -= Tbot * gamma3[N3+1]
#
# Solución del sistema lineal
#
T3[1:N3+1] = np.linalg.solve(C2,f3)
#

# Para N2 = 110 nodos
# Aplicación de las condiciones de frontera tipo Dirichlet.
#
T4[0] = Ttop
T4[N4+1] = Tbot 
f4[0] -= Ttop * gamma4[0]
f4[N4-1] -= Tbot * gamma4[N4+1]
#
# Solución del sistema lineal
#
T4[1:N4+1] = np.linalg.solve(D2,f4)
```

```python
#
# GRAFICACIÓN
#
ax1 = plt.subplot(121)
plt.plot(T,z1, 'r-', lw=2.0, label='FDM.', zorder=10, c = 'g')
plt.plot(T2,z2, 'r-', lw=2.0, label='FDM N=95', zorder=10, c= 'y')
plt.plot(T3,z3, 'r-', lw=2.0, label='FDM N=102', zorder=10, c= 'b')
plt.plot(T4,z4, 'r-', lw=2.0, label='FDM N=110', zorder=10, c= 'r')
plt.plot([T[0], T[-1]], [z1[0], z1[-1]], '--', lw=0.75, c='k' )
plt.ylim(4100,-100)
plt.xlim(-10,800)
plt.title('Solución Numérica')
plt.xlabel('$T$ [$^o$C]')
plt.ylabel('Profundidad [m]')
plt.minorticks_on()
plt.legend()
plt.grid()

ax2 = plt.subplot(122, sharey=ax1)
plt.plot(A, z1, '.--', lw=0.75, c='k')
plt.xlabel('$D_{th}$ [m$^2$/s]')
plt.xlim(0.65e-6,1.05e-6)
plt.minorticks_on()
plt.ticklabel_format(axis='x', style='sci', scilimits=(0,0))
plt.title('Coeficiente $D_{th}$')
plt.grid()

plt.subplots_adjust(hspace=1.0)

plt.show()
```
